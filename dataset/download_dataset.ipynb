```python
from datasets import load_dataset
from pathlib import Path

def download_imagenet(save_dir="./imagenet"):    
    save_dir = Path(save_dir)
    
    print("Downloading ImageNet-1K from HuggingFace...")
    print(f"Save location: {save_dir}")
    
    print("Downloading training set (1.28M images)...")
    train_dataset = load_dataset(
        "imagenet-1k",
        split="train",
        cache_dir=str(save_dir / "cache")
    )
    
    print("\nDownloading validation set (50K images)...")
    val_dataset = load_dataset(
        "imagenet-1k",
        split="validation",
        cache_dir=str(save_dir / "cache")
    )
    
    print("\nDownload complete!")
    print(f"Training: {len(train_dataset):,} images")
    print(f"Validation: {len(val_dataset):,} images")
    print(f"Location: {save_dir}")
    
    return train_dataset, val_dataset
    
if __name__ == "__main__":
    download_imagenet("/project2/jieyuz_1727/snehansh/imagenet")
```
